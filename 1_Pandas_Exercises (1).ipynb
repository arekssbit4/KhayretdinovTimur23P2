19. What is the price of each item?

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
for item in chipo['item_name'].unique():
    item_rows = chipo[chipo['item_name'] == item]
    most_common_price = item_rows['item_price'].mode()[0]
    print(f"{item}\t{most_common_price:.2f}")
    
```

20. Sort by the name of the item

```python
if chipo['item_price'].dtype == 'object':
    chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)

unique_items = sorted(chipo['item_name'].unique())

for item in unique_items:
    rows = chipo[chipo['item_name'] == item]
    most_common_price = rows['item_price'].mode()[0]
    print(f"{item}\t{most_common_price:.2f}")
    
```
